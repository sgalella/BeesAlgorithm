```python
import sys
sys.path.append('..')

import numpy as np
import bees_algorithm as ba
```

```python
resolution = 100
limits = [-5, 5, -3, 3]  # x_min, x_max, y_min, y_max
num_iterations = 20
landscape = ba.SphereLandscape(limits, resolution)
```

```python
np.random.seed(1234)
bees = ba.BeesAlgorithm(landscape, n=20, m=3, e=1, nep=10, nsp=7, ngh=0.5)
ba.VisualizeSearch.show_last(bees, num_iterations)
```

## Show iterative search

```python
%matplotlib
np.random.seed(1234)
bees = ba.BeesAlgorithm(landscape, n=20, m=3, e=1, nep=10, nsp=7, ngh=0.5)
ba.VisualizeSearch.show_all(bees, num_iterations)
```

## Plot best fitness achievedÂ¶

```python
np.random.seed(1234)
bees = ba.BeesAlgorithm(landscape, n=20, m=3, e=1, nep=10, nsp=7, ngh=0.5)
for _ in range(num_iterations):
    recruiters_e, best_e, recruiters_m_e, best_m_e = bees.recruit_scouts()
    bees.abandon_locations(recruiters_e, best_e, recruiters_m_e, best_m_e)
print(bees.best_fitness)
```
